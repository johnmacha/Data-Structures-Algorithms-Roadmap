```python
from collections import deque
class Stack():
    def __init__(self):
        self.container = deque()
    def add(self, val):
        self.container.append(val)
```

```python
s = Stack()

text = "We will conquer COVID-19"

for char in text: #To reverse a string using a stack, you must push each character separately.
    s.add(char)

s.container.reverse()
# s.container
print(''.join(s.container))
```
