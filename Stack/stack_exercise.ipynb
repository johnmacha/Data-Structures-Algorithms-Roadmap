```python
# method 1
from collections import deque
class Stack():
    def __init__(self):
        self.container = deque()
    def add(self, val):
        self.container.append(val)
```

```python
s = Stack()

text = "We will conquer COVID-19"

for char in text: #To reverse a string using a stack, you must push each character separately.
    s.add(char)

s.container.reverse()
# s.container
print(''.join(s.container)) #So Python will join everything without spaces between them.
```

```python
# method 2
from collections import deque
class Letters():
    def __init__(self):
        self.phrase = deque()
    
    def push(self, val):
        self.phrase.append(val)
    
    def pop(self):
        return self.phrase.pop()

    def peek(self):
        return self.phrase[-1]
    
    def is_empty(self):
        return len(self.phrase) == 0
    
    def size(self):
        return len(self.phrase)
```

```python
def reverse_string(s):
    letter = Letters()

    for ch in s:
        letter.push(ch)
    
    rstr = ''
    while letter.size()!=0:
        rstr += letter.pop()
    
    return rstr
```

```python
print(reverse_string("We will conquer COVID-19"))
```

```python
print(reverse_string("Liverpool is the best!"))
```
