```python
# 2. Write a function in python that checks if parenthesis in the string are balanced or not. Possible parantheses are "{}',"()" or "[]". Use Stack class from the tutorial.
# is_balanced("({a+b})")     --> True
# is_balanced("))((a+b}{")   --> False
# is_balanced("((a+b))")     --> True
# is_balanced("))")          --> False
# is_balanced("[a+b]*(x+2y)*{gg+kk}") --> True
```

```python
from collections import deque
class Stack():
    def __init__(self):
        self.items = deque()
    
    def push(self, val):
        self.items.append(val)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
    
    def peek(self):
        return self.items[-1]
    
    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

```python
def is_balanced(s):
    stack = Stack()
    pairs = {
        '}':'{',
        ')':'(',
        ']':'['
    }
    for char in s: #This loops through every character in the string
        if char in "([{":#This checks if the character is an opening bracket
           stack.push(char)
        #(for top devs)
        #if char in pairs.values
        #'check' print (pairs)

        elif char in ")]}":#This checks for closing brackets
            if stack.is_empty():#If there's no bracket to match 
                return False
            #if char in pairs - makes code more scalable
            #'check' print char
            
            if stack.pop() != pairs[char]: #(i)pop the last opening bracket (ii)Find the expected opening bracket
                return False
    return stack.is_empty() #ensures no unmatched opening brackets remain
```

```python
print(is_balanced("(({a+b}))"))
```
