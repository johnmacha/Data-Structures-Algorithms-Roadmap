```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range (self.MAX)]

    def hash_map(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX
        
```

```python
t = HashTable()
t.hash_map('december 15')
```
