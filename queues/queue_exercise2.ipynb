```python
# Write a program to print binary numbers from 1 to 10 using Queue. Use Queue class implemented in main tutorial.
# You also need to add front() function in queue class that can return the front element in the queue.
# Hint: After 1 second and third number is 1+0 and 1+1. 4th and 5th number are second number (i.e. 10) + 0 and second number (i.e. 10) + 1.
```

```python
# Method 1
from collections import deque

class Queue():
    def __init__(self):
        self.my_queue = deque()
    
    def add(self, val):
        self.my_queue.append(val)
    
    def popitem(self):
        return self.my_queue.popleft()
    
    def front(self):
        return self.my_queue[0] # just peek don't remove
    
    def empty(self):
        return len(self.my_queue)==0
    
    def size(self):
        return len(self.my_queue)
```

```python
# Using Breadth First Search(BFS) approach
def my_binary(b):
    q = Queue()
    results = []

    q.add("1") #Start with 1 in the queue

    for r in range(b):
        #Take front element(s) add to the results
        current = q.popitem()
        results.append(current)
        #Push both back to the queue
        q.add(current + "0")
        q.add(current + "1")

    return results
```

```python
my_binary(10)
```
