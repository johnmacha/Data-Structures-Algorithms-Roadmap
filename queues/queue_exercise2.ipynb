```python
from collections import deque

class Queue():
    def __init__(self):
        self.my_queue = deque()
    
    def add(self, val):
        self.my_queue(val)
    
    def pop(self):
        return self.my_queue.pop()
    
    def empty(self):
        return len(self.my_queue)==0
    
    def size(self):
        return len(self.my_queue)
```
