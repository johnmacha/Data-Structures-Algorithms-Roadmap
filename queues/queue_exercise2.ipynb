```python
from collections import deque
import time

class Queue():
    def __init__(self):
        self.my_queue = deque()
    
    def add(self, val):
        self.my_queue.append(val)
    
    def popitem(self):
        return self.my_queue.pop()
    
    def front(self):
        return self.popitem()
    
    def empty(self):
        return len(self.my_queue)==0
    
    def size(self):
        return len(self.my_queue)
```

```python
q = Queue()

def binary(b):
    for m in b:
       q.append(1)
       if q.front() == 1:
           q.append(1) + 0
       else:
            q.append(m) + 1
    # q.my_queue
```
