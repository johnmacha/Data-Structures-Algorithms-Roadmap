```python
# Getting the maximum temperature in the first 10 days of Jan
class HashTable():
    def __init__(self):
        self.MAX = 20
        self.arr = [[] for i in range(self.MAX)]
    
    def gethash(self, key):
        h = 0

        for char in key:
            h += ord(char)
        
        return h % self.MAX
    
    def __setitem__(self, key, val):
        h = self.gethash(key)

        for idx, element in enumerate(self.arr[h]):
            if element[0] == key:
                self.arr[h][idx] = (key, val)
                return
        
        self.arr[h].append((key, val))
    
    def __getitem__(self, key):
        h = self.gethash(key)

        for element in self.arr[h]:
            if element[0] == key:
                return element[1]
        return None
```

```python
nyc_weather = HashTable()
with open('./nyc_weather.csv', 'r') as f:
    next(f)
    for line in f:
        date, temperature = line.strip().split(',')
        nyc_weather[date] = int(temperature)
```

```python
days = ["Jan 1", "Jan 2", "Jan 3", "Jan 4", "Jan 5", "Jan 6", "Jan 7","Jan 8", "Jan 9", "Jan 10"]

# traversal for maximum temperature
maximum_temp = float("-inf") # '-inf' is negative infinity to get the maximum
for day in days:
    temp = nyc_weather[day]
    if temp > maximum_temp:
        maximum_temp = temp 
```
