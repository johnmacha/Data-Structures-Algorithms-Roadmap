```python
#checking the output of the csv file
nyc_weather = []
with open("./nyc_weather.csv", "r") as f:
     for line in f: tokens = line.strip().split(',') 
     date = tokens[0] 
     temperature = tokens[1]
     nyc_weather.append([date, temperature])
```

```python
# Getting the average temperature for the first week of Jan
class HashTable():
    def __init__(self):
        self.MAX = 20
        self.arr = [[] for i in range(self.MAX)]
    
    def gethash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX
    
    def __setitem__(self, key, val):
        h = self.gethash(key)
        for idx, element in enumerate(self.arr[h]):
            if element[0]==key:
                self.arr[h][idx] = (key, val)
                return

        self.arr[h].append((key, val))
    
    def __getitem__(self, key):
        h = self.gethash(key)
        for element in self.arr[h]:
            if element[0] == key:
                return element[1]
        return None
      


```

```python
nyc_weather = HashTable() #creating object for HashTable
with open("./nyc_weather.csv", "r") as f:
    next(f) #skip header
    for line in f:
        date, temperature = line.strip().split(',')
        nyc_weather[date] = int(temperature)
```

```python
days = ["Jan 1", "Jan 2", "Jan 3", "Jan 4", "Jan 5", "Jan 6", "Jan 7"]

total = 0
for day in days:
    total += nyc_weather[day]

average_temp = total / len(days)
```
