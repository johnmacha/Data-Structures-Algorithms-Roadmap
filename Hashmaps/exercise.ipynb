```python
# Getting the average temperature for the first week of Jan
class HashTable():
    def __init__(self):
        self.MAX = 20
        self.arr = (None for i in range(self.MAX))
    
    def gethash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX
    
    # traversal for getting average temperature
    # traversal for the first 7 days (0-7 index)
    def average(self, key):
        h = self.gethash(key)

        for element in self.arr[h]:
            if element[0] == key:
                avg = element[1] / len(self.arr[h])
        return avg
        
  
```
