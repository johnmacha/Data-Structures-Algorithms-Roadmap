```python
class HashTable:
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range (self.MAX)]

    def get_hash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX

    def __setitem__(self, key, val): #__setitem__ and __getitem__ are 'dunder methods'
        h = self.get_hash(key)
        self.arr[h] = val

    def __getitem__(self, key):
        h = self.get_hash(key)
        return self.arr[h]

    def __delitem__(self, key): #This deletes an item
        h = self.get_hash(key)
        self.arr[h] = None
        
```

```python
t = HashTable()
t.get_hash('december 15')
t['march 6'] = 120
t['march 1'] = 20
t['dec 17'] = 80
```
